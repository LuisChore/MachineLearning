# Line Chart
Display any kind of 1D signal: time series, sound waves, etc

```
import numpy as np 
import matplotlib.pyplot as plt

X = np.linspace(0,20,1000)
Y = np.sin(X) + 0.2*X
plt.plot(X,Y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function');
```

# Scatterplot
Geometric perspective of our data, classification or regression models in a geometric way, with very high dimention, we first try to reduce the dimentions to see the data in one plot

```
X = np.random.randn(100,2) # 100 samples, 2 attributes
plt.scatter(X[:,0],X[:,1]);
```

```
X = np.random.randn(200,2)
X[0:100] = X[0:100] + 3 # classification problem
plt.scatter(X[:,0],X[:,1]);
```

```
Y = np.zeros(200) # its required the labes to change the color
Y[0:100] = 1;
plt.scatter(X[:,0],X[:,1],c = Y);
```

# Histogram
To see the probability distribution of our data

```
X = np.random.randn(10000)
plt.hist(X);
```

```
# adding more bars(bins)
plt.hist(X,bins = 50);
```
